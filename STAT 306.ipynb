```r
library(tidyverse)
library(broom)
```

```r
property <- read.csv2(
  file = "property-tax-report.csv"
) |>
  subset(select = c(CURRENT_LAND_VALUE,
    PREVIOUS_LAND_VALUE,
    PREVIOUS_IMPROVEMENT_VALUE,
    YEAR_BUILT
  )) |>
  na.omit()

  property2 <- subset(property,
  is.finite(CURRENT_LAND_VALUE) & CURRENT_LAND_VALUE > 0 &
  is.finite(PREVIOUS_LAND_VALUE) & PREVIOUS_LAND_VALUE > 0 &
  is.finite(PREVIOUS_IMPROVEMENT_VALUE) & PREVIOUS_IMPROVEMENT_VALUE > 0 &
  is.finite(YEAR_BUILT)
)

set.seed(306)               
n <- 2000

property_sample <- property2[sample.int(nrow(property2), size = min(n, nrow(property2))), ]

write.csv(property_sample, "property_sample_2000.csv", row.names = FALSE)

```

```r
fit <- lm(log(CURRENT_LAND_VALUE) ~ log(PREVIOUS_LAND_VALUE) + log(PREVIOUS_IMPROVEMENT_VALUE) + YEAR_BUILT, data = property2)
bb <- coef(fit)
```
